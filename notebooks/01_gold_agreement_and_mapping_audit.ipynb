# Gold construction, agreement, and mapping audit

Verifies corpus counts, raw applicability agreement, exact-consensus subsets,
and the audited 3-class intensity mapping.

```python
from pathlib import Path

def find_repo_root():
    current = Path.cwd().resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "data" / "gold").exists() and (candidate / "results").exists():
            return candidate
    raise FileNotFoundError(
        "Repository root not found. Run the notebook from inside the repository."
    )

ROOT = find_repo_root()

import pandas as pd
from sklearn.metrics import cohen_kappa_score
GOLD = ROOT / 'data' / 'gold'
RESULTS = ROOT / 'results' / 'tables'
app = pd.read_csv(GOLD / 'applicability_gold_final_1497.csv')
pairwise = pd.read_csv(GOLD / 'final_pairwise_annotations_1497.csv')
axis3 = pd.read_csv(GOLD / 'axis_gold_consensus_3class_only.csv')
axis5 = pd.read_csv(GOLD / 'axis_gold_consensus_5class_only.csv')
print('Total:', len(app))
print('Applicable:', int((app['applicability_score'] == 1).sum()))
print('Excluded:', int((app['applicability_score'] == 0).sum()))
print('3-class consensus:', len(axis3))
print('5-class consensus:', len(axis5))
```

```python
a1 = next(c for c in pairwise.columns if 'a1' in c.lower() and 'applic' in c.lower())
a2 = next(c for c in pairwise.columns if 'a2' in c.lower() and 'applic' in c.lower())
valid = pairwise[[a1, a2]].dropna().astype(int)
print('Exact agreement:', (valid[a1] == valid[a2]).mean())
print('Cohen kappa:', cohen_kappa_score(valid[a1], valid[a2], labels=[0, 1]))
print('Conflicts:', int((valid[a1] != valid[a2]).sum()))
```

# External 2024 audit and matrices

Verifies the temporal projection, pipeline agreement, matrix counts, and the
corpus-wide linguistic audit.

```python
from pathlib import Path

def find_repo_root():
    current = Path.cwd().resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "data" / "gold").exists() and (candidate / "results").exists():
            return candidate
    raise FileNotFoundError(
        "Repository root not found. Run the notebook from inside the repository."
    )

ROOT = find_repo_root()

import pandas as pd
DATA = ROOT / 'data' / 'external_2024'
RESULTS = ROOT / 'results' / 'tables'
corpus = pd.read_csv(DATA / 'corpus_2024.csv')
print('Records:', len(corpus))
print('Unique hashes:', corpus['text_hash'].nunique())
display(pd.read_csv(RESULTS / '2024_pipeline_agreement.csv'))
display(pd.read_csv(RESULTS / '2024_region_counts.csv'))
```
